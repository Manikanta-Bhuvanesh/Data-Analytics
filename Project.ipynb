## **Filling null values with mean of respective column**

```python
df['bmi'].fillna(df['bmi'].mean(),inplace=True)
```

## **Droping id column**

```python
df.drop('id',axis=1,inplace=True)
print(df.head())
```

## **Printing unique elements in each column**

```python
print(df.smoking_status.unique())
print(df.gender.unique())
print(df.hypertension.unique())
print(df.heart_disease.unique())
print(df.ever_married.unique())
print(df.Residence_type.unique())
print(df.work_type.unique())
```

## **En-coding the columns**

```python
enc=LabelEncoder()
gender=enc.fit_transform(df['gender'])
smoking_status=enc.fit_transform(df['smoking_status'])
work_type=enc.fit_transform(df['work_type'])
Residence_type=enc.fit_transform(df['Residence_type'])
ever_married=enc.fit_transform(df['ever_married'])
print(gender,smoking_status,work_type,Residence_type,ever_married)
```

```python
df['ever_married']=ever_married
df['Residence_type']=Residence_type
df['smoking_status']=smoking_status
df['gender']=gender
df['work_type']=work_type
```

```python
df.head()
```

```python
print(df.smoking_status.unique())
print(df.gender.unique())
print(df.hypertension.unique())
print(df.heart_disease.unique())
print(df.ever_married.unique())
print(df.Residence_type.unique())
print(df.work_type.unique())
```

## **Normalizing columns**

```python
scaler=MinMaxScaler()
attributes=["age","avg_glucose_level","bmi"]
df[attributes] = scaler.fit_transform(df[attributes])
```

```python
df.head()
```

## **Assigning the input and ouptut**

```python
X=df.drop('stroke',axis=1)
print(X)

Y=df['stroke']
print(Y)
```

## **Splitting the input and output to training and testing sets**

```python

X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.2)
X_train
```

# **Model building**

```python
#importing the required libraries
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import Dense

model = Sequential()
lyrs=[5]
model.add(Dense(lyrs[0], input_dim=X_train.shape[1], activation='linear'))
for i in range(1,len(lyrs)):
    model.add(Dense(lyrs[i], activation='linear'))
model.add(Dropout(0.2))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='Adam', metrics=['accuracy'])
```

# **Model Fitting**

```python
t=model.fit(X_train, Y_train, epochs=50, batch_size=32,validation_split=0.2, verbose=0)
```

# **Prediction**

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
my_predict=[]
for i in model.predict(X_test):
    if i>0.5:
        my_predict.append(1)
    if i<0.5:
        my_predict.append(0)
accuracy= accuracy_score(Y_test,my_predict)
accuracy*100
```

```python
plt.plot(t.history['accuracy'])
plt.plot(t.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```
